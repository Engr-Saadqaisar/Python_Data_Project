```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
# Load the dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleaning
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
df.loc[:10,"salary_year_avg":"salary_hour_avg"]
```

```python
df.info()
```

```python
median_salary_yearly= df["salary_year_avg"].median()
median_salary_hourly= df["salary_hour_avg"].median()
df_filled=df
df_filled["salary_hour_avg"]=df_filled["salary_hour_avg"].fillna(median_salary_hourly)
df_filled["salary_year_avg"]=df_filled["salary_year_avg"].fillna(median_salary_yearly)
df_filled.loc[:10,"salary_year_avg":"salary_hour_avg"]
```

```python
df.loc[:10,"salary_year_avg":"salary_hour_avg"]
```

```python
# Removing Duplicates
print("Lenght of the dataset before removing duplicates: ",len(df_filled))
df_unique=df_filled
df_unique=df_unique.drop_duplicates()
print("Lenght of the dataset after removing duplicates: ",len(df_unique))
print("Rows removed: ",len(df_filled)-len(df_unique))
```

```python
df_unique=df_unique.drop_duplicates(subset=['job_title','company_name'])
print("Lenght of the dataset before removing duplicates: ",len(df_filled))
print("Lenght of the dataset after removing duplicates: ",len(df_unique))
print("Rows removed: ",len(df_filled)-len(df_unique))
```
