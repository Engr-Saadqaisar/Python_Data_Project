```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
# Load the dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleaning
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
```

```python
df_cleaned=df[pd.notna(df['salary_year_avg'])].copy()

def protected_salary(salary):
    return salary*1.03

df_cleaned['salary_year_inflated'] = df_cleaned['salary_year_avg'].apply(protected_salary)

df_cleaned[['salary_year_inflated', 'salary_year_avg']].head()
```

```python
def clean_skills(skills):
    if pd.isna(skills):
        return []
    else:
        return ast.literal_eval(skills)

df_cleaned['job_skills_cleaned'] = df_cleaned['job_skills'].apply(clean_skills)


df_cleaned['job_skills'].apply(lambda x: [] if pd.isna(x) else ast.literal_eval(x)).head()

```

Make another column named projected salary 2, where I will assume that senior roles will get 5 percent increase and others will get 3 percent.
----------------------------------------------------------------------------------------------

```python
def projected_salary_2(row):
    if pd.isna(row['job_title_short']) or pd.isna(row['job_skills']):
        return []
    else:    
        
        if 'senior' in row['job_title_short'].lower():
            salary = row['salary_year_avg']
            return salary * 1.05

        else:
            salary = row['salary_year_avg']
            return salary * 1.03

df_cleaned['salary_year_inflated_2'] = df_cleaned.apply(projected_salary_2, axis=1)

df_cleaned[['salary_year_inflated_2', 'salary_year_avg', 'job_title_short', 'salary_year_inflated']].sample(10)
```
